```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import publiplots as pp

pp.set_notebook_style()
```

```python
eval_table = pd.read_csv("~/Downloads/evaluation_results.csv")
freqs = ((eval_table=="YES").sum()/len(eval_table))[3:]
freqs_df = pd.DataFrame({"rate":freqs})
freqs_df.index = freqs_df.index.str.split(' - ', n=1, expand=True) 
freqs_df = freqs_df.reset_index(names=["model","Model Configuration"])
freqs_df
```

```python
fig, ax = pp.barplot(data=freqs_df,
                     x="model",y="rate", hue="Model Configuration",
                     title="Performance accross LLM models",
                     xlabel="Model Name", ylabel="Pass Rate")
plt.xticks(rotation=45, ha="right")
#plt.legend(title="Model Configuration")
```
